```python
# 1)

# Juego simple de piedra, papel o tijeras (ppt)

import random

def ppt():
    opciones = ["piedra", "papel", "tijeras"]
    comp = random.choice(opciones)

    # Entrada del usuario validada (convertimos a minpusculas para validar)
    while True:
        jugador = input("Elige piedra, papel o tijeras: ").lower()
        if jugador in opciones:
            break
        print("Entrada inválida. Intenta de nuevo.")

    print(f"Tú: {jugador} | Computadora: {comp}")

    # Reglas del juego
    if jugador == comp:
        print("Resultado: Empate")
    elif (jugador == "piedra" and comp == "tijeras") or \
         (jugador == "papel" and comp == "piedra") or \
         (jugador == "tijeras" and comp == "papel"):
        print("Resultado: Ganas =)")
    else:
        print("Resultado: Pierdes =()")

# Llamada a función

ppt()
```

```python
# 2)

# Simulación del juego de dados entre Juan y María.
# Reglas:
# Si sale 1, se repite el turno
# Si sale 6, empate
# Si sale 2 o 3, gana Juan
# Si sale 4 o 5, gana María

import random

while True:
    dado = random.randint(1, 6)
    print(f"\nResultado del dado: {dado}")

    if dado == 1:
        print("Sale 1: Se repite el lanzamiento.")
        continue
    elif dado == 6:
        print("Sale 6: Empate.")
    elif dado in (2, 3):
        print("Juan gana.")
    elif dado in (4, 5):
        print("María gana.")
    break  # termina el juego después de resolver el resultado
```
