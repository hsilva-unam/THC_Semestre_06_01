```python
# PROBLEMA 1

# Función para evaluar el acceso a la bóveda
def acceso_boveda(edad_juan, edad_pedro, num_nietos):
    # Condición a: Juan es mayor de 35 o Pedro menor de 50
    condicion_a = edad_juan > 35 or edad_pedro < 50

    # Condición b: la suma de edades es un número par
    condicion_b = (edad_juan + edad_pedro) % 2 == 0

    # Condición c: Manuel tiene al menos 1 nieto
    condicion_c = num_nietos >= 1

    # Evaluamos si todas las condiciones se cumplen
    if condicion_a and condicion_b and condicion_c:
        return "SI - Acceso"
    else:
        return "NO - No hay acceso"


# --- Programa principal ---
juan = int(input("Edad de Juan: "))
pedro = int(input("Edad de Pedro: "))
nietos = int(input("Número de nietos de Manuel: "))

resultado = acceso_boveda(juan, pedro, nietos)
print("\n¿Tienen acceso a la bóveda?:", resultado)
```

```python
# PROBLEMA 3

# Función que evalua si un numero es primo

def es_primo(n):

#    Regresa verdadero si n es primo, falso en caso contrario.
#    Considera n >= 2

    if n == 2:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True


# ---- Programa principal ----
while True:
    try:
        num = int(input("Ingresa un entero mayor que 1: "))
        if num <= 1:
            print("Error en el Número")
            continue
        print(es_primo(num))   # imprime True o False
        break
    except ValueError:
        print("Entrada inválida: escribe un entero.")
```

```python
# PROBLEMA 4

def procesa_numero():
    while True:
        try:
            N = int(input("Ingresa un entero mayor a 100: "))

            # Regla 1
            if N <= 100:
                print("Error en el número")
                continue

            ultimo = N % 10

            # Regla 2
            if ultimo in (4, 7, 9):
                if N % 3 == 0:
                    print(f"{N} es divisible por 3.")
                else:
                    print(f"{N} NO es divisible por 3.")
                return None

            # Regla 3
            elif ultimo in (1, 5):
                resultado = N // 10
                print(f"{N}//10 = {resultado}")
                return resultado

            # Regla 4
            else:
                residuo = N % 3
                print(f"Residuo de {N} ÷ 3 = {residuo}")
                return residuo

        except ValueError:
            print("Entrada inválida: escribe un entero.")

# Llamada a la función
procesa_numero()
```

```python
# PROBLEMA 5


# MCD y MCM

def mcd(a, b):
    a, b = abs(a), abs(b)
    while b != 0:
        a, b = b, a % b
    return a

def mcm(a, b):
    if a == 0 or b == 0:
        return 0
    return abs(a * b) // mcd(a, b)

# Programa principa
while True:
    try:
        a = int(input("a: "))
        b = int(input("b: "))
        resultados = {"a": a, "b": b, "MCD": mcd(a, b), "MCM": mcm(a, b)}
        for k, v in resultados.items():
            print(f"{k}: {v}")
        break
    except ValueError:
        print("Entrada inválida: escribe enteros.")

```

```python
# PROBLEMA 6

# Separa una lista mixta en cuatro listas por tipo: int, str, bool y float.

def filtrar_por_tipo(datos):
    enteros, cadenas, booleanos, flotantes = [], [], [], []
    for x in datos:
        if isinstance(x, bool):
            booleanos.append(x)
        elif isinstance(x, int):
            enteros.append(x)
        elif isinstance(x, float):
            flotantes.append(x)
        elif isinstance(x, str):
            cadenas.append(x)
    return enteros, cadenas, booleanos, flotantes

# --- Datos del enunciado ---
datos = [25, "Matemáticas", 3.141592, True, 8, False, "Nombres", 2.718, "UNAM",
         0, 9.81, "CdMx", 7e12, 12.5, "2025", True, "población", 0.01, 42, False]

enteros, cadenas, booleanos, flotantes = filtrar_por_tipo(datos)

print("Enteros  :", enteros)
print("Cadenas  :", cadenas)
print("Booleanos:", booleanos)
print("Flotantes:", flotantes)
```

```python
# PROBLEMA 7

# Calcula la distancia (en km) entre dos puntos geográficos usando
# la fórmula de Haversine

import math

R = 6378.1  # radio aproximado de la Tierra en km

def grados_a_radianes(grados):
    return grados * math.pi / 180

def haversin(angulo):
    return math.sin(angulo / 2) ** 2

def distancia_haversine(lat1, lon1, lat2, lon2):
    # Convertir coordenadas a radianes
    lat1, lon1 = grados_a_radianes(lat1), grados_a_radianes(lon1)
    lat2, lon2 = grados_a_radianes(lat2), grados_a_radianes(lon2)

    # Aplicar fórmula
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = haversin(dlat) + math.cos(lat1) * math.cos(lat2) * haversin(dlon)
    d = 2 * R * math.asin(math.sqrt(a))
    return d


# Pruebas
pares = [
    ("París", 48.9, 2.4, "Roma", 41.9, 12.5),
    ("CDMX", 19.43, -99.13, "Tokio", 35.68, 139.69),
    ("CDMX", 19.43, -99.13, "Los Ángeles", 34.05, -118.24),
]

for ciudad1, lat1, lon1, ciudad2, lat2, lon2 in pares:
    d = distancia_haversine(lat1, lon1, lat2, lon2)
    print(f"Distancia {ciudad1} → {ciudad2}: {d:.2f} km")
```

```python
# PROBLEMA 8

# Imprime la serie de Fibonacci hasta n

def fibonacci(n):
    a, b = 0, 1
    serie = []
    for _ in range(n):
        serie.append(a)
        a, b = b, a + b
    return serie


#  Programa principal
while True:
    try:
        n = int(input("Número de términos de la serie de Fibonacci: "))
        if n <= 0:
            print("Error: ingresa un entero positivo.")
            continue
        print(fibonacci(n))
        break
    except ValueError:
        print("Entrada inválida: escribe un entero.")
```

```python
# PROBLEMA 9

# Procesa listas de sismos para:
# 1) Convertir magnitudes a float
# 2) Capitalizar nombres de zonas
# 3) Clasificar intensidad con etiquetas
# 4) Contar eventos por zona
# 5) Mostrar los 3 sismos más fuertes en formato solicitado


#  Datos proporcionados
fechas = ["2025-03-01", "2025-03-02", "2025-03-03", "2025-03-04", "2025-03-05",
          "2025-03-06", "2025-03-07", "2025-03-08", "2025-03-09", "2025-03-10"]

zonas = ["guerrero", "oaxaca", "chiapas", "michoacán", "guerrero",
         "oaxaca", "chiapas", "guerrero", "oaxaca", "chiapas"]

magnitudes = ["3.8", "4.2", "5.0", "2.9", "4.8", "3.5", "4.1", "5.3", "5.0", "3.0"]

prof_km = [15, 80, 12, 5, 35, 120, 60, 25, 10, 90]  # profundidad en km


# 1. Convertir magnitudes
magnitudes = list(map(float, magnitudes))

# 2. Capitalizar nombres de zona (lambda + map)
zonas = list(map(lambda z: z.capitalize(), zonas))

# 3. Crear lista intensidad
def clasifica(m):
    if m < 3.5:
        return "Muy Leve"
    elif m < 4.5:
        return "Leve"
    elif m < 5.5:
        return "Moderado"
    else:
        return "Fuerte"

intensidad = list(map(clasifica, magnitudes))

# 4. Diccionario con conteo por zona
conteo_por_zona = {}
for z in zonas:
    conteo_por_zona[z] = conteo_por_zona.get(z, 0) + 1

# 5. Ordenar por magnitud descendente y mostrar los 3 mayores
# Empaquetamos todos los datos para poder ordenarlos juntos
eventos = list(zip(fechas, zonas, magnitudes, prof_km))

# Ordenar por magnitud (índice 2) en orden descendente
eventos_ordenados = sorted(eventos, key=lambda x: x[2], reverse=True)

print("Conteo por zona:")
for zona, cant in conteo_por_zona.items():
    print(f"{zona}: {cant}")

print("\n 3 sismos más fuertes:")
for fecha, zona, mag, prof in eventos_ordenados[:3]:
    print(f"{fecha} | {zona} | {mag} | {prof} km")
```

```python
# PROBLEMA 10

# Realiza productos escalar y vectorial entre tres vectores a, b, c
# indica si algunos son paralelos o perpendiculares


# Vectores iniciales
a = [2, -1, 4]
b = [4, -2, 8]
c = [-5, 2, 1]


def dot(u, v):
    """Producto escalar (u·v)."""
    return sum(ui * vi for ui, vi in zip(u, v))

def cross(u, v):
    """Producto vectorial (u x v) para vectores 3D."""
    return [
        u[1]*v[2] - u[2]*v[1],
        u[2]*v[0] - u[0]*v[2],
        u[0]*v[1] - u[1]*v[0]
    ]

def es_perpendicular(u, v):
    """Dos vectores son perpendiculares si su producto escalar es 0."""
    return dot(u, v) == 0

def es_paralelo(u, v):
    """Dos vectores son paralelos si uno es múltiplo escalar del otro."""
    try:
        razones = [ui / vi for ui, vi in zip(u, v) if vi != 0]
        return len(set(razones)) == 1
    except ZeroDivisionError:
        return False


# Producto escalar
print("Producto escalar:")
print("a · b =", dot(a, b))
print("b · c =", dot(b, c))
print("c · a =", dot(c, a))

# Producto vectorial
print("\n Producto vectorial:")
print("a × b =", cross(a, b))
print("b × c =", cross(b, c))
print("a × c =", cross(a, c))

# Paralelos / perpendiculares
print("\n Relaciones entre vectores:")

pares = [(a, b, "a y b"), (b, c, "b y c"), (a, c, "a y c")]

for v1, v2, nombre in pares:
    if es_perpendicular(v1, v2):
        print(f"{nombre} son perpendiculares")
    elif es_paralelo(v1, v2):
        print(f"{nombre} son paralelos")
    else:
        print(f"{nombre} no son ni paralelos ni perpendiculares")
```
